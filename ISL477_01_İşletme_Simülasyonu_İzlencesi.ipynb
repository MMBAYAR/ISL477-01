###**Dersin Tanımı**

###**Ölçme-Değerlendirme Sistemi**
Derste öğrenci faaliyetleri grup çalışması içinde yürütülecektir. Grup içinde akran değerlendirmesi, oyunda verilen kararların dayanak ve gerekçe raporları ve şirket (oyun) performansı bireysel notlamaya esas oluşturacaktır.
* 6 Düşünme şapkası rolleri ve kararların akademik gerekçelendirme raporları her karar başlığı için en çok 1 sayfa (Standart yazı tipleri 1.5 satır aralığı)olacak biçimde haftalık olarak hazırlanacaktır. **(%80)**
* Şirket performansı haftalık olarak simülasyon uygulamasında 5 yıldız ile temsil olunmaktadır. **(%20)**<br><br>

**NOT:** Grupların en düşük şirket performansı gösterdiği 1 period not değerlendirmesinde dikkate alınmayacaktır.<br><br>
